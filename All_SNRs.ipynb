```python
from fastai import *
from fastai.vision import *
import numpy as np
import os
path = os.getcwd() + '/training_zone/2016_rgb'
path = Path(path)
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train= 'train', valid_pct=0.2, num_workers=8, bs=128).normalize()
learn = cnn_learner(data, models.resnet18, metrics=[accuracy, error_rate])
```

```python
SNR = 18
learn.load("stage-3")
test = Path(os.getcwd() + "/training_zone/2016_rgb/test/%s"%str(SNR))
data_ = ImageDataBunch.from_folder(test, valid_pct = 0.9, num_workers=8).normalize()
learn.data = data_
learn.fit_one_cycle(1)
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
##accuracy  = 91.4%
```
